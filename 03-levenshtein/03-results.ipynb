```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from bokeh.models import FactorRange
```

```python
results = pd.read_csv("data/results2015.csv", header=None, names=['word', 'counter'])
results.head(10)
```

```python
def plotHistogram(values, min_val, max_val, bins, show_info=False, logarithmic=True):
    events, edges, patches = plt.hist(values, range=(min_val, max_val), bins=bins, log=logarithmic)
    if show_info:
        print("Elements matched:")
        print(events)
        print("Bins boundaries:")
        print(edges)
        print("Patches:")
        for p in patches:
            print(p)

    plt.xlabel('single word occurances in all judgements (range [%.2f, %.2f])' % (min_val, max_val), 
               fontsize=15, 
               labelpad=20)
    plt.ylabel('number of words', fontsize=15)
    plt.title('Words occurances in judgments (2015) histogram')
    plt.rcParams["figure.figsize"] = [14,7]
    plt.rcParams.update({'font.size': 17})
    plt.xticks(edges, rotation=90)
    plt.rcParams['xtick.labelsize'] = 12
    plt.rcParams['ytick.labelsize'] = 12
    plt.rcParams['figure.figsize'] = 20, 10
    plt.show()

all_values = results['counter']
plotHistogram(all_values, min(all_values), max(all_values), 30)
```
